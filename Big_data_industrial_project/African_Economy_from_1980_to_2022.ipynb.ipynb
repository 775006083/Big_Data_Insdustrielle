# **5) étude comparative sur l'inflation pré-covid, pendant covid et post-covid pour les pays formant l’UEMOA.**

```python
#Aficher les 5 pays de l'Afrique de l'Ouest
UEMOA_countries = data[data['Country'].isin([ 'Benin', 'Burkina Faso', 'Guinea-Bissau', "Cote d'Ivoire", 'Mali', 'Niger', 'Senegal','Togo'])]                 
```

```python
UEMOA_countries.head()
```

```python
sns.set_theme(style="darkgrid")
sns.lineplot(x="Gross domestic product, (constant prices US$)", y="Year",
             hue="Country",
             data=UEMOA_countries)
```

```python
UEMOA_2020 = UEMOA_countries.loc[UEMOA_countries['Year']==2020]
```

```python
UEMOA_2020.groupby(by = ['Country', 'Year'])["Gross domestic product, (constant prices US$)"].value_counts().plot.pie(subplots=True,figsize=(7,7),autopct='%2.1f%%')
```

# Afficher l'Histogramme des PIB des pays Africains qui s'etant de l'année 2015 à 2019

```python
UEMOA_2015_2019 = UEMOA_countries.loc[(UEMOA_countries['Year']>=2015) & (UEMOA_countries['Year']<=2019)]
```

```python

plt.figure(figsize=(30,10))
ax = sns.barplot(x='Country', y= "Gross domestic product, (constant prices US$)",hue='Year',data=UEMOA_2015_2019,ci=None)
ax.set_title('Gross domestic product, (constant prices US$) et Year')
plt.show()
```

# Afficher l'Histogramme des PIB des pays Africains qui s'etant de l'année 2021 à 2022

```python
UEMOA_2021_2022 = UEMOA_countries.loc[(UEMOA_countries['Year']>=2021) & (UEMOA_countries['Year']<=2022)]
```

```python

plt.figure(figsize=(30,10))
ax = sns.barplot(x='Country', y= "Gross domestic product, (constant prices US$)",hue='Year',data=UEMOA_2021_2022,ci=None)
ax.set_title('Gross domestic product, (constant prices US$) et Year')
plt.show()
```
