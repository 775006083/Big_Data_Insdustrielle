# import data with libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../input/african-economy-from-1980-to-2022/ObservationData_lavlqce.csv")
```

```python
data.head()
```

# data parameters exploration

```python
data.columns
```

```python
data.dtypes
```

```python
# see the data that have nan values
data[data.isna().any(axis = 1)].head(len(data[data.isna().any(axis = 1)]))
```

```python
data_21st = data[data['Year'] >= 2000]
```

```python
data_21st.fillna(method='ffill' ,inplace=True)
```

```python
data_21st.isna().sum()
```

# Try to get the important insights

```python
##get average real GDP grwoth in 21st century by country and by year
country_high = pd.pivot_table(data_21st , values = ['Real GDP growth (annual %)'] , index = ['Country'] , aggfunc = np.mean)
country_high.sort_values(by = 'Real GDP growth (annual %)' , ascending = False , inplace = True)
```

```python
plt.figure(figsize = (16,16))
plt.xticks(rotation = 90)
sns.barplot(x =country_high.index , y = 'Real GDP growth (annual %)' , data = country_high )
plt.show()
```

```python
# as we see in th above plot that Ethiopia have a high Real GDP growth (annual %) in africa
# and south sudan one in africa
## calculate the total import and export of goods and service in dollar for each country
#'Exports of goods and services (current US$)',
#'Imports of goods and services (current US$)'
total_import_export = pd.pivot_table(values = ['Exports of goods and services (current US$)',
'Imports of goods and services (current US$)'],
index='Country' , data = data_21st , aggfunc = np.sum)
```

```python
plt.figure(figsize = (16,16))
plt.xticks(rotation = 90)
plt.plot(total_import_export)
plt.xlabel('Countries')
plt.legend(['Exports of goods and services (current US$)',
'Imports of goods and services (current US$)'] ,loc='upper left')
```

```python
mini = np.min(total_import_export['Exports of goods and services (current US$)'])
print(total_import_export[total_import_export['Exports of goods and services (current US$)'] == mini])
mini = np.min(total_import_export['Imports of goods and services (current US$)'])
print(total_import_export[total_import_export['Imports of goods and services (current US$)'] == mini])
```

```python
maxi = np.max(total_import_export['Exports of goods and services (current US$)'])
print(total_import_export[total_import_export['Exports of goods and services (current US$)'] == maxi])
maxi = np.max(total_import_export['Imports of goods and services (current US$)'])
print(total_import_export[total_import_export['Imports of goods and services (current US$)'] == maxi])
```

##### as we see that Sao Tome and Principe have a low exports and imports of goods in africa in this century
##### and South Africa have high exports and imports of goods in africa in this century

```python
#'Gross capital formation, Private sector (current US$)',
#'Gross capital formation, Public sector (current US$)'
avg_private_public = pd.pivot_table(values = ['Gross capital formation, Private sector  (current US$)',
'Gross capital formation, Public sector  (current US$)'],
index='Country' , data = data_21st , aggfunc = np.mean)
```

```python
plt.figure(figsize = (16,16))
plt.xticks(rotation = 90)
plt.plot(avg_private_public)
plt.xlabel('Countries')
plt.legend(['Gross capital formation, Private sector (current US$)',
'Gross capital formation, Public sector (current US$)'] ,loc='upper center')
```

```python
#'Central government, Fiscal Balance (Current US $)',
#'Central government, total revenue and grants (Current US $)',
#'Central government, total expenditure and net lending (Current US $)',
# now we can get everage of these titles for all country
avg_central_dovernment = pd.pivot_table(values = ['Central government, Fiscal Balance (Current US $)',
'Central government, total revenue and grants  (Current US $)',
'Central government, total expenditure and net lending  (Current US $)'],
index='Country' , data = data_21st , aggfunc = np.mean)
```

```python
plt.figure(figsize = (16,16))
plt.xticks(rotation = 90)
plt.plot(avg_central_dovernment)
plt.xlabel('Countries')
plt.legend(['Central government, Fiscal Balance (Current US $)',
'Central government, total revenue and grants (Current US $)',
'Central government, total expenditure and net lending (Current US $)'] ,loc='upper center')
```

#### let's explore the highest and lowest year in Real GDP growth in egypt to see the difference between them

```python
max_gdp = egypt['Real GDP growth (annual %)'].max()
min_gdp = egypt['Real GDP growth (annual %)'].min()
```

```python
high = egypt[egypt['Real GDP growth (annual %)'] == max_gdp]
```

```python
low = egypt[egypt['Real GDP growth (annual %)'] == min_gdp]
```

```python
high_low_data = high.append(low) 
high_low_data
```

> As we see that the highest real GDP growth for Egypt was in 1983 and the lowest rate in 2011 , I think that this is because of the revolution

> let's compare between the high and low with some feature

```python
## compare by the General government and Household final consumption expenditure (current US$)
fig, ax = plt.subplots(1,2,figsize=[16,5])
x = high_low_data.Year
sns.barplot(x =x , y = high_low_data['General government final consumption expenditure (current US$)'],ax = ax[0]);
# ax[0].title("")
sns.barplot(x =x , y = high_low_data['Household final consumption expenditure (current US$)'] , ax = ax[1]);
fig.suptitle("final consumption expenditure (current US$)");
```

```python
# Gross capital formation, Private sector  (current US$)                   
# Gross capital formation, Public sector  (current US$)
x = high_low_data.Year
fig, ax = plt.subplots(1,2,figsize=[16,5])
x = high_low_data.Year
sns.barplot(x =x , y = high_low_data['Gross capital formation, Private sector  (current US$)'],ax = ax[0]);
# ax[0].title("")
sns.barplot(x =x , y = high_low_data['Gross capital formation, Public sector  (current US$)'] , ax = ax[1]);
fig.suptitle("final consumption expenditure (current US$)");
```
